### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
#code modified from tutorial to fit by needs
class myCNN(nn.Module):
  def __init__(self, name = "myCNN"):
          super(myCNN, self).__init__()
          self.conv1 = nn.Conv2d(in_channels=3, out_channels=5, kernel_size=5) # input 3 because RGB, kernel size is 5 (stride is 1)
          #input 3*224*224 output is (output=5)*(224-(kernel=5 - 1))*(224-(kernel=5 - 1)) ergo 5*220*220
          self.pool = nn.MaxPool2d(kernel_size=2, stride=2) 
          # First reduction: to reduce dimentionality by half on each width and height ergo 5*110*110 is output
          # Second reduction: to reduce dimentionality by half on each width and height ergo 10*53*53 is output (this is the input to the first fully connected layer)
          self.conv2 = nn.Conv2d(in_channels=5, out_channels=10, kernel_size=5) # input 5 because of previous output, kernel size is 5 (stride is 1)
          #input 5*110*110 output is (output=10)*(110-(kernel=5 - 1))*(110-(kernel=5 - 1)) ergo 10*106*106
          self.fc1 = nn.Linear(28090, 32) #32 to make sure is divisible by two
          #10*53*53 is input for first fully connected
          self.fc2 = nn.Linear(32, 9) #to classify
#Note:  I tried using different kernel sizes but I couldnt make them work appropiately
  def forward(self, x):
          x = self.pool(F.relu(self.conv1(x))) #Activation, first pool  set
          x = self.pool(F.relu(self.conv2(x))) #Activation, second pool  set
          x = x.view(-1, 28090)
          x = F.relu(self.fc1(x)) 
          x = self.fc2(x)
          return x
```

```python
#code modified from tutorial to fit by needs
#second CNN with a modified kernel number on the 2nd convolutional layer; to be used in part 3
class myCNN1(nn.Module):
  def __init__(self, name = "myCNN1"):
          super(myCNN1, self).__init__()
          self.conv1 = nn.Conv2d(in_channels=3, out_channels=5, kernel_size=5) #in 3*224*224 out 5*220*220

          self.pool = nn.MaxPool2d(kernel_size=2, stride=2) 
          #in1 5*220*220 out 5*110*110
          #in2 10*106*106 out 10*53*53
          self.conv2 = nn.Conv2d(in_channels=5, out_channels=10, kernel_size=5) #in 5*110*110 out 10*106*106

          self.conv3 = nn.Conv2d(in_channels=10, out_channels=20, kernel_size=5) #in 10*53*53 out 20*49*49
          
          self.fc1 = nn.Linear(48020, 32) #32 to make sure is divisible by two
          #10*53*53 is input for first fully connected
          self.fc2 = nn.Linear(32, 9) #to classify
#Note:  I tried using different kernel sizes but I couldnt make them work appropiately
  def forward(self, x):
          x = self.pool(F.relu(self.conv1(x))) #Activation, first pool  set
          x = self.pool(F.relu(self.conv2(x))) #Activation, second pool  set
          x = F.relu(self.conv3(x))
          x = x.view(-1, 48020)
          x = F.relu(self.fc1(x)) 
          x = self.fc2(x)
          return x
```

### Part (b) Training Code - 5 pt

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

```python
#torch.manual_seed(981) # set seed
#I chose CE because of the nature of the problem (classification)
# I chose Adam for the optimizer because of the fact that I have already seen its benefits in ECE421 vs SGD, its pretty efficient
#Shamelessly taken from tutorial (minor modifications )
def get_accuracy(model, loader):

    correct = 0
    total = 0
    for imgs, labels in loader:

        #############################################
        #To Enable GPU Usage
        if use_cuda and torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()
        #############################################

        output = model(imgs)
        
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
    return correct / total

def train(model, training_dataset,validation_dataset, batches=64, num_epochs=1, lr = 0.001):
    #To use batch sizes we need to set them and prepare a DataLoader for each of the sets
    #torch.manual_seed(981)
    training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=batches, shuffle=True) #Shuffle
    validation_loader = torch.utils.data.DataLoader(validation_dataset, batch_size=batches,  shuffle=True)
    

    criterion = nn.CrossEntropyLoss() 
    optimizer = optim.Adam(model.parameters(), lr=lr) 
    testing_acc = 0.
    train_acc, val_acc = [], []
    iterations = 0
    # training
    n = 0 # the number of iterations
    plot_epochs = 0
    for epoch in range(num_epochs):
        for imgs, labels in iter(training_loader):
          
          
            #############################################
            #To Enable GPU Usage
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()
            #############################################
            
              
            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch
            iterations =iterations+1

        plot_epochs = plot_epochs +1
        # save the current training information
        train_acc.append(get_accuracy(model, training_loader)) # compute training accuracy 
        val_acc.append(get_accuracy(model, validation_loader))  # compute validation accuracy
        #print("Epoch ", epoch+1, "training accuracy: ",train_acc[epoch])
        #print("Epoch ", epoch+1, "validation accuracy: ",val_acc[epoch])

    #testing_acc = get_accuracy(model, testing_loader)
    # plotting
    plot_epochs_l = np.arange(1, plot_epochs + 1)
    print("Epochs: {0}, Batch Size: {1}, Learning Rate: {2}".format(plot_epochs, batches, lr))
    plt.title("Training Curve")
    plt.plot(plot_epochs_l, train_acc, label="Train")
    plt.xlabel("Epochs")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()

     
    plt.title("Validation Curve")
    plt.plot(plot_epochs_l, val_acc, label="Validation")
    plt.xlabel("Epochs")
    plt.ylabel("Validation Accuracy")
    plt.legend(loc='best')
    plt.show()

    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))
    
    print("Final Epochs: {}".format(plot_epochs))
    print("Final Iterations: {}".format(iterations))

    return train_acc, val_acc
```

### Part (c) “Overfit” to a Small Dataset - 5 pt

One way to sanity check our neural network model and training code is to check whether the model is capable
of “overfitting” or “memorizing” a small dataset. A properly constructed CNN with correct training code
should be able to memorize the answers to a small number of images quickly.

Construct a small dataset (e.g. just the images that you have collected). Then show that your model and
training code is capable of memorizing the labels of this small data set.

With a large batch size (e.g. the entire small dataset) and learning rate that is not too high, You should be
able to obtain a 100% training accuracy on that small dataset relatively quickly (within 200 iterations).

```python
def overfit(model, small_dataset, num_epochs=50, lr = 0.001):
    #torch.manual_seed(981)
    small_loader = torch.utils.data.DataLoader(small_dataset, batch_size=27,  shuffle=True)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    train_acc = []
    plot_epochs = 0
    iterations = 0
    for epoch in range(num_epochs):
        for imgs, labels in iter(small_loader):
            #GPU
            if use_cuda and torch.cuda.is_available():
              imgs = imgs.cuda()
              labels = labels.cuda()

            out = model(imgs)             # forward pass
            loss = criterion(out, labels) # compute the total loss
            loss.backward()               # backward pass (compute parameter updates)
            optimizer.step()              # make the updates for each parameter
            optimizer.zero_grad()         # a clean up step for PyTorch
            iterations = iterations+1

        # Acc
        train_acc.append(get_accuracy(model, small_loader))
        print("Epoch: {0}, Accuracy: {1}".format(epoch+1, train_acc[-1]))
        plot_epochs = plot_epochs +1
    
    plot_epochs_l = np.arange(1, plot_epochs + 1)
    print("Epochs: {0}, Batch Size: {1}, Learning Rate: {2}".format(plot_epochs, '27', lr))
    plt.title("Training Curve")
    plt.plot(plot_epochs_l, train_acc, label="Train")
    plt.xlabel("Epochs")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()
    print("Final Training Accuracy: {}".format(train_acc[-1]))
    
    print("Final Epochs: {}".format(plot_epochs))
    print("Final Iterations: {}".format(iterations))

    return train_acc
```
